```python
import pandas as pd

df1 = pd.read_csv('spider_1_1600.csv')
df2 = pd.read_csv('spider_1601_3200.csv')
df3 = pd.read_csv('spider_3201_4800.csv')
df4 = pd.read_csv('spider_4801_6400.csv')
df5 = pd.read_csv('spider_6401_8034.csv')
```

```python
result = pd.concat([df1, df2, df3, df4, df5])
```

```python
result = result.sort_values('Unnamed: 0')
```

```python
result = result.reset_index().drop('index', axis=1)
```

```python
result.to_csv('spider_250.csv')
```

```python
duplicates = result.duplicated(keep=False)
duplicate_rows = result[duplicates]
occurrence_count = duplicate_rows.groupby(duplicate_rows.columns.tolist()).size().reset_index(name='counts')
more_than_twice = occurrence_count[occurrence_count['counts'] > 1]
```
