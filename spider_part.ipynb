```python
import pandas as pd

df_full = pd.read_csv('spider.csv')
df = df_full.loc[1600:3200, ['Unnamed: 0', 'db_id', 'query', 'question', 'SELECT', 'JOIN', 'GROUP BY', 'HAVING', 'ORDER BY']]

#df.to_csv('spider_1600.csv')

drop_condition1 = (df['SELECT'] == 1) & (df['JOIN'] == 0) & (df['GROUP BY'] == 0) & (df['HAVING'] == 0)
drop_condition2 = (df['SELECT'] == 1) & (df['JOIN'] == 0) & (df['GROUP BY'] == 1)
drop_condition3 = (df['SELECT'] == 1) & (df['JOIN'] == 1) & (df['GROUP BY'] == 0)

df_578 = df.drop(df.index[drop_condition1 | drop_condition2 | drop_condition3])

# drop duplicated queries
df_365 = df_578.drop_duplicates(subset = 'query')
```
