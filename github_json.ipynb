### tables.json in spider github

```python
import pandas as pd
import json

with open("tables.json", "r") as f:
    tbl_data = json.load(f)
    
tbl_df = pd.DataFrame(tbl_data)

df_sorted = tbl_df.sort_values('db_id').reset_index().drop('index', axis = 1)

df_final = df_sorted.loc[:,['db_id', 'table_names_original', 'column_names_original', 'column_types', 'primary_keys', 'foreign_keys']].rename(columns={'table_names_original': 'table_names', 'column_names_original': 'column_names'})
```

### DDL dataframe initializing (columns = [db_id, table_name])

```python
with open("tables.json", "r") as f:
    data = json.load(f)
    
spider_selected = pd.read_csv('SPIDER_248.csv')
db_lst = spider_selected['db_id'].unique()
```

```python
df = pd.DataFrame(columns=['db_id', 'table_name'])

for db_id in db_lst:
    for info in data:
        if info['db_id'] == db_id:
            db_id = info['db_id']
            table_names = info['table_names_original']
            
            tmp = pd.DataFrame({'db_id': [db_id] * len(table_names), 'table_name': table_names})
            df = pd.concat([df, tmp], ignore_index=True)
```
