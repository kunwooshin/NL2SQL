```python
import pandas as pd
import json

with open("dev.json", "r") as f:
    data = json.load(f)
    
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
components = df.iloc[0, -1].keys()
for comp in components:
    column_name = 'sql_' + comp
    df[column_name] = df.iloc[:, -1].apply(lambda x: x[comp] if isinstance(x, dict) and comp in x else None)

df = df.drop('sql', axis = 1)

```
